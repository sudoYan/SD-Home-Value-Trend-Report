## Zillow Home Value Index Dataset

In our ZHVI dataset, we note that there are several neighborhoods for which data was collected only after a certain point in time (after which they should have consistent observations). We do not want to remove neighborhoods that are missing values prior to this "start date" because they will still be useful for comparisons over the course of their respective tenures. However, some neighborhoods may have inconsistent or missing values after this first observation. We will be looking into removing such rows.

```python
import pandas as pd
import numpy as np
#ZHVI Dataset
zhvi_df = pd.read_csv("sd_zhvi_2000_to_2025_nhbd.csv")
zhvi_df.drop(['RegionType', 'StateName', 'State', 'City', 'Metro', 'CountyName'], axis=1, inplace=True)
zhvi_df.head()
```

```python
zhvi_cols = zhvi_df.columns[3:]
obs_started = zhvi_df[zhvi_cols].notnull().cumsum(axis=1) > 0
incomp_obs = zhvi_df[zhvi_cols].isnull() & obs_started
nbhds_incomp = incomp_obs.any(axis=1)
problem_rows = zhvi_df.loc[nbhds_incomp]
rows_clean = ~incomp_obs.any(axis=1)
clean_rows_df = zhvi_df.loc[rows_clean]
clean_rows_df.head(5)
```

However, we will only be using data on a quarterly basis for the housing price discussion due to the granularity of the transit stop data. Therefore, we will enact a rule that ensures that datasets with a minimum of one value per non-overlapping quarter of a year will be retained in the data while those who do not meet this expectation will be removed. This ensures the data is useful for our analysis.

```python
def quarter_nonnull(row):
    first_valid = row.first_valid_index()
    if first_valid is None:
        return False

    start = row.index.get_loc(first_valid) + 1
    tb_complete = row.iloc[start:]

    for i in range(0, len(tb_complete), 3):
        group = tb_complete.iloc[i:i + 3]
        if group.notnull().sum() == 0:
            return False 
    return True

mask_valids = zhvi_df[zhvi_cols].apply(quarter_nonnull, axis=1)
clean_zhvi = zhvi_df.loc[mask_valids]
clean_zhvi.head(5)
```

This shows that the clean_zhvi dataset is clean according to the quarter-based non-null rule. This is the dataset ready for use in this project

### 1. Checking the Seasonality of ZHVI

A significant issue with the usage of housing market data is the effect of seasonality on the market. This has the potential to considerably skew our data and confound our findings, since of much of the housing variance could potentially be explained at a seasonal level. That is why this analysis is important to conduct in other to ensure that we do not make assumptions about the smoothness of our data. First, we will look at the data in the ZHVI dataset.

```python
zhvi_df = clean_zhvi
zhvi_df.head()
```

Now, we do a cursory look at the ZHVI across the neighborhoods through a scatterplot across time within a year for every year. This is to look visibly at whether seasonal trends impact the housing market in some way. We will make the graphs into a GIF to allow for ease of viewing. This is merely a preview of how seasonal vs overall trends are shaping up by looking at the year-on-year trends from 2000 to 2025 (till March). 

Note: It may be necessary to use a Jupyter Notebook-style application to view the GIF (it may not display on GitHub). For ease of viewing, we have also attached a link to view the GIF in case the application is not available.

```python
from PIL import Image
import io
from IPython.display import display, Image as IPyImage

def seasonal_check(year):
    month_columns = [col for col in zhvi_df.columns if year in col]
    zhvi2000_long = zhvi_df[month_columns].melt(var_name='Month', value_name='Price')
    zhvi2000_long['Month'] = pd.to_datetime(zhvi2000_long['Month'], format='%Y-%m-%d')

    sns.scatterplot(data=zhvi2000_long, x='Month', y='Price', alpha=0.4, color='gray')
    sns.lineplot(data=zhvi2000_long, x='Month', y='Price', errorbar=None, estimator='mean', color='magenta', label='Average Trend')

    plt.title(f'House Prices in All Neighborhoods ({year})')
    plt.xlabel('Month')
    plt.ylabel('House Price')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.legend()

frames = []
for year in range(2000, 2026):
    seasonal_check(str(year))
    buf = io.BytesIO()
    plt.savefig(buf, format='png')
    plt.close()
    buf.seek(0)
    frames.append(Image.open(buf))

gif_buffer = io.BytesIO()
frames[0].save(gif_buffer, format='GIF', append_images=frames[1:], save_all=True, duration=2000, loop=0)
gif_buffer.seek(0)

display(IPyImage(data=gif_buffer.getvalue()))
```

It does not appear that seasonal trends contributes to ZHVI variation. We have to note that some neighborhoods only had ZHVI data from a particular point in time. In addition, some neighborhoods have missing values erratically. We will create a heatmap-style visualization to show variation in house prices is present over both axes. As we can see from below, there seems to be a strong smoothing effect from the ZHVI's data collection. However, some regions to show up with minor month-based crests and troughs within the boundaries of their overall year-on-year trends. We can see faint differences for July-August (peaks) and March-April (troughs).

```python
df_anova = zhvi_df[zhvi_df['2000-01-31'].notnull()]
df_anova = df_anova.melt(var_name='Date', value_name='Price', value_vars=zhvi_df.columns.difference(['RegionID', 'SizeRank', 'RegionName']))
df_anova['Date'] = pd.to_datetime(df_anova['Date'], format='%Y-%m-%d')
df_anova['Year'] = df_anova['Date'].dt.year
df_anova['Month'] = df_anova['Date'].dt.month_name()
df_anova['Month'] = pd.Categorical(df_anova['Month'], categories=['January', 'February', 'March','April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December'], ordered=True)
df_anova = df_anova.dropna(subset=['Price'])

heatmap_data = df_anova.groupby(['Year', 'Month'])['Price'].mean().unstack()
plt.figure(figsize=(12, 8))
sns.heatmap(heatmap_data, cmap='coolwarm', annot=False, fmt=".0f", linewidths=0.5)

plt.title('Monthly House Price (ZHVI) Trends by Year')
plt.xlabel('Month')
plt.ylabel('House Price')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

In order to find out conclusively the differences in the explanation of variance by month and year, we will conduct an ANOVA test. We do this because this analysis is being done over multiple variable comparisons. To do this, we will use statsmodels.api. For a variable to be statistically significant, its F-statistic must have a p-value smaller than 0.05. Furthermore, we will 

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols
model = ols('Price ~ C(Month) + C(Year)', df_anova).fit()
anova_table = sm.stats.anova_lm(model, typ=2)

anova_table
```

```python
ss_total = anova_table.loc['C(Month)', 'sum_sq'] + anova_table.loc['C(Year)', 'sum_sq'] + anova_table.loc['Residual', 'sum_sq']
eta_sq_month = anova_table.loc['C(Month)', 'sum_sq'] / ss_total
eta_sq_year = anova_table.loc['C(Year)', 'sum_sq'] / ss_total
eta_sq_res = anova_table.loc['Residual', 'sum_sq'] / ss_total
print("Results of the eta-squared test")
print(f"η² for C(Month): {eta_sq_month}")
print(f"η² for C(Year): {eta_sq_year}")
print(f"η² for Residual (Not Explained): {eta_sq_res}")
```

This model tells us that while Year explains almost all of the explained variance in house prices (showing that overall price level of houses drive changes in ZHVI), there is a small but statistically significant amount (p < 0.05) explained by Month. This shows that the ZHVI Index (despite being smoothed, but not specifically to San Diego) still has some level of its variance explained by month-based changes. The eta-squared test for the model shows us that C(Month) explains only 0.08% of the variance while C(Year) explains 38.75% of the variance. 

This is important because if we are to use these datapoints to explain the effect of transit on data, we must be aware that such changes could confound the effect of transit if not accounted for and rectified. However, our results show us that 0.08% of the variance in our model here is explained by Month-on-month differences. This allows us to adapt our assumptions about how we will conduct our future housing price analysis.

```python
import pandas as pd
import numpy as np
import geopandas as gpd
from shapely.geometry import Point
import matplotlib.pyplot as plt
```

```python
transit_gdf = gpd.GeoDataFrame(transit_df, geometry=gpd.points_from_xy(y=transit_df.lat, x=transit_df.lng), crs="EPSG:4326")
transit_gdf.head(5)
```

For geocoding the neighborhoods of San Diego, we use the map used by the San Diego Police Department which assigns beats to neighborhoods. This encoding is saved to neighborhoods. We have plotted it using the shapefile in the Github.

```python
neighborhoods = gpd.read_file("SDPD_Beats_shapefile/SDPD_Beats.shp")
neighborhoods.plot(figsize=(10, 10), color="yellow", edgecolor="black")
plt.title("Neighborhoods of San Diego", fontsize=12)
plt.axis("off")
plt.show()
```

However, some parts of the San Diego city are military zones, agricultural land, or natural preserves. These are areas which often cannot have transit stations built within and often around. These serve to obfuscate our results as they do not show the effect of transit on housing, but rather the effect of various confounding factors like the prevalence of agricultural land or military regulation. Thus, we will be removing these regions from our neighborhoods and analysis by using a Zoning Map of San Diego. The Zoning Codes AR-1-1 and AG-1-1 correspond to these artificially restricted zones. We have kept AR-1-2 on the grounds that the minimum area of housing is 1 acre which allows for greater residential density.

```python
zones = gpd.read_file("Zoning_Base_SD_shapefile/Zoning_Base_SD.shp")
uncounted_zones = zones[zones["ZONE_NAME"].isin(["AR-1-1", "AG-1-1", "AR-1-2"])]
uncounted_zones.plot(figsize=(6, 6), color="green", edgecolor="black")
plt.title("Zoning Areas with Code AR-1-1/AG-1-1", fontsize=12)
plt.axis("off")
plt.show()
```

```python
#Ensure and define the coordinate system being used for the geopandas dataframes
transit_gdf = transit_gdf.to_crs(epsg=26911)
neighborhoods = neighborhoods.to_crs(epsg=26911)
uncounted_zones = uncounted_zones.to_crs(epsg=26911)
```

Now we will remove these specific Zoned Areas from the Neighborhoods of San Diego.

```python
neighborhoods_cleaned = gpd.overlay(neighborhoods, uncounted_zones, how='difference')
neighborhoods_cleaned.plot(figsize=(6, 6), color="green", edgecolor="black")
plt.title("Neighborhoods without Code AR 1-1", fontsize=16)
plt.axis("off")
plt.show()
```

Now, there are a few ways to understand transit density based on the granularity and precision of our analysis. First, we will conduct a neighborhood-based analysis where we utilize a centroid of 850 m for neighborhoods to find out the general transit stops within 850 meters (the buffer) which is equivalent to 10 minutes of walking. Now, there are specific ways to get exact walking time limits to bus stops for a particular neighborhood. However, these practices are power-consuming, inefficient and do not serve the purpose of illuminating transit density any greater than analysis through centroid-based granularity. Thus, it is a caveat of this analysis that the buffer is based more on distance from neighborhood centroids which is in part to work with the granularity of the ZHVI dataset.

```python
neighborhoods_cleaned["buffer_850m"] = neighborhoods_cleaned.geometry.centroid.buffer(850)
buffered = gpd.GeoDataFrame(neighborhoods_cleaned, geometry="buffer_850m")
joined = gpd.sjoin(transit_gdf, buffered, predicate="within", how="inner")
stop_counts = joined.groupby("NAME").size().reset_index(name="stop_count_850m")
neighborhoods_cleaned = neighborhoods_cleaned.merge(stop_counts, on="NAME", how="left")
neighborhoods_cleaned["stop_count_850m"] = neighborhoods_cleaned["stop_count_850m"].fillna(0)
```

Let us take a look at neighborhoods that have 0 stops within 850 meters of their centroids. This allows us to understand neighborhoods that may be a part of our control groups for housing price analysis since they lack transit access (within adequate walking distance).

```python
neighborhoods_cleaned[neighborhoods_cleaned['stop_count_850m'] == 0]
```

Finally, let us make a map of the San Diego region which shows us transit density (number of stops within an 850 meter radius from the centroid for each neighborhood).

```python
neighborhoods_cleaned.plot(column="stop_count_850m", cmap="plasma", legend=True, figsize=(10,10))
plt.title("Transit stops within 850m distance (Centroid)", fontsize=16)
plt.axis("off")
plt.show()
```

As you can see, the transit density analysis by this particular metric shows us intuitive notions where certain regions such as Downtown show a better concentration of transit stops than some others like the Scripps Ranch neighborhood. However, the center of a neighborhood, especially a large one, is not necessarily the center of population and residential/commercial/industrial life (which is where transit is required and provided). Therefore, utilizing zones as a means of understanding population and residential/commerical/industrial distributions within a neighborhoods allows us to more precisely score the neighborhoods on their transit access. Now, in the next section of the analysis (the conclusion), we will use centroid analysis on zones. However, these zones can vary dramatically in size. That is why we need to remove overlaps between zones within a neighborhood to prevent deduplication from skewing the data. Finally, this map will represent a more accurate score based on the number of zones serviced by a particular station and the number of stations within the 850m centroid of each zone.

```python
zones_c = zones[~zones["ZONE_NAME"].isin(["AR-1-1", "AG-1-1", "AR-1-2"])]
zones_c['zone_id'] = zones_c.index.astype(str)

transit_gdf = transit_gdf.to_crs(epsg=26911)
neighborhoods_cleaned = neighborhoods_cleaned.to_crs(epsg=26911)
zones_c = zones_c.to_crs(epsg=26911)

zones_c['centroid'] = zones_c.geometry.centroid
zones_centroids = zones_c.set_geometry('centroid')
zones_centroids['buffer_850m'] = zones_centroids.geometry.buffer(850)
buffers = zones_centroids.set_geometry('buffer_850m')

stops_with_buffers = gpd.sjoin(transit_gdf, buffers, predicate='within', how='left')
stops_with_buffers = stops_with_buffers.rename(columns={'zone_id': 'zone_id'})
zone_centroids = zones_centroids.set_geometry('centroid')
zones_with_nhoods = gpd.sjoin(zone_centroids, neighborhoods_cleaned, predicate='within', how='left')[['zone_id', 'NAME']]
zones_c = zones_c.merge(zones_with_nhoods, on='zone_id', how='left')

stops_with_buffers = stops_with_buffers.merge(zones_c[['zone_id', 'NAME']], on='zone_id', how='left')
stop_zone_counts = stops_with_buffers.groupby(['objectid', 'NAME']).size().reset_index(name='n_zone_buffers')
stops_with_weights = stops_with_buffers.merge(stop_zone_counts, on=['objectid', 'NAME'], how='left')
stops_with_weights['weighted_contribution'] = 1 / stops_with_weights['n_zone_buffers']

zone_scores = stops_with_weights.groupby(['zone_id', 'NAME'])['weighted_contribution'].sum().reset_index()
neighborhood_scores = zone_scores.groupby('NAME')['weighted_contribution'].sum().reset_index(name='neighborhood_score')

neighborhoods_cleaned = neighborhoods_cleaned.merge(neighborhood_scores, on='NAME', how='left').fillna({'neighborhood_score': 0})
neighborhoods_cleaned.plot(column='neighborhood_score', cmap='plasma', legend=True, figsize=(12,12))
plt.title("Transit Stop Score (10 min walk adjusted)", fontsize=16)
plt.axis("off")
plt.show()
```

Here, the map resembles the previous one at its peaks, but neighborhoods previously considered to have a transit-stop-score of 0 are now much more nuanced. This is important as it shows us the depth and contrast in these neighborhoods. However, there are still neighborhoods that have significantly lower transit stop scores, which will be useful in constructing the control and experimental groups for our analysis.

## FINAL ANALYSIS - HOW MUCH DOES TRANSIT ACCESS CONTRIBUTE TO HOME VALUE IN SAN DIEGO?

Now that our exploratory analyses have shown us how certain neighborhoods have better transit density than others, we will tease out the quantifiable effect of transit access on Home Values in San Diego (represented by ZHVI). This will allow us to generate inferences and conclusions on what transit access means for the housing market in San Diego. We will use methods such as gravity-weighting with distance decay, difference-in-differences estimation and size-weighted WLS analysis for this.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path
import geopandas as gpd
import statsmodels.formula.api as smf
import seaborn as sns
```

First, we get our neighborhoods dataset and transit_gdfs dictionary ready by clearing it up as well as converting it into a GeoDataFrame.

Now that we have a cleaned neighborhoods dataset without the Agricultural areas (as explained above and in the previous analyses), we will now calculate our gravity scores. These scores will also be weighted by the type of station. Trolley stations will receive a weight of 3:1 compared to bus stations based on the specific layout of San Diego wherein trolley stations provide service at a frequency approximately 3 times that of an average bus station - this is including bus, rapid bus, and light rail connections. Given this, the gravity score is based on provision of service with each stop contributing an inverse-distance score to the model itself.

```python
def compute_gravity_scores(transit_gdfs, neighborhoods_gdf, radius=1500, weight_map=None):
    weight_map = weight_map or {0: 1, 1: 3}
    gravity_scores = pd.DataFrame()
    
    # Ensure projection is consistent (use local projection like EPSG:2230)
    neighborhoods = neighborhoods_gdf.copy()
    neighborhoods = neighborhoods.to_crs(epsg=2230)
    neighborhoods['centroid'] = neighborhoods.geometry.centroid
    neighborhoods = neighborhoods.set_index('neighborhood')
    
    for yymm, transit_gdf in transit_gdfs.items():
        print(f"Processing {yymm}...")
        transit_gdf = transit_gdf.to_crs(neighborhoods.crs)
        
        # Assign weights to stops
        transit_gdf['weight'] = transit_gdf['location_type'].map(weight_map).fillna(1)

        scores = []
        for idx, hood in neighborhoods.iterrows():
            center = hood['centroid']
            score = 0
            for _, stop in transit_gdf.iterrows():
                dist = center.distance(stop.geometry)
                if 0 < dist <= radius:
                    score += stop['weight'] / (dist ** 2)
            scores.append(score)
        
        gravity_scores[yymm] = scores

    gravity_scores.index = neighborhoods.index
    return gravity_scores

gravity_df = compute_gravity_scores(transit_gdfs, neighborhoods_cleaned)
gravity_df
```

Now, we will create 2 columns for identifying neighborhoods that have no change in their transit scores and those that have no transit access (if applicable). This is useful for looking into the data and developing a reasonable line of application for our hypothesis.

```python
def num_cols(row):
    yymm_cols = []
    for col in list(row.index):
        if str(col).isdigit() == True:
            yymm_cols.append(col)
    yymm_cols = sorted(yymm_cols, key=int)
    return row[yymm_cols]

def grp_sorter(row):
    yymm_row = num_cols(row)
    constant = yymm_row.iloc[7]
    for yymm in yymm_row:
        if constant != yymm:
            return 0
    return 1

def trans_sorter(row):
    if row['no_change'] == 0:
        return 0
    else:
        row = num_cols(row)
        if row.iloc[0] == 0:
            return 1
        else:
            return 0
gravity_df['no_change'] = gravity_df.apply(grp_sorter, axis=1)
gravity_df['no_transit'] = gravity_df.apply(trans_sorter, axis=1)
gravity_df = gravity_df.merge(neighborhoods_cleaned[['neighborhood', 'geometry']], on='neighborhood', how='left')
#gravity_df.to_csv('gravity_df.csv', index = True)
gravity_df = gpd.GeoDataFrame(gravity_df, geometry='geometry', crs='EPSG:2230')
gravity_df
```

Now, let us create a column that shows score differences and at which observation scores are changing within neighborhoods. Then, we will create a subset for treatment effects where this score_diff is not 0. This will be the set of observations we will eventually use for our analysis and defining our pre- and post- treatment periods.

```python
zhvi_transit = zhvi_transit[zhvi_transit['ZHVI'] > 0].copy()
zhvi_transit['score_diff'] = zhvi_transit.groupby('neighborhood')['trans_score'].diff().copy()
zhvi_transit['score_diff'] = zhvi_transit['score_diff'].where(zhvi_transit['score_diff'].abs() > 1e-12, 0).copy()
```

Here we define a function that will help us define the pre- and post-treatment period by moving between observations in the formatting of our column names - YYMM.

```python
def shift_yymm(yymm, months):
    year = yymm // 100
    month = yymm % 100
    total_months = year * 12 + month + months
    new_year = total_months // 12
    new_month = total_months % 12
    if new_month == 0:
        new_year -= 1
        new_month = 12
    return new_year * 100 + new_month
```

Now we create the treatment_events subset and also the log of our ZHVI. While ZHVI alone would show us absolute trends, this is not as important to our analysis. Having access to percentage (log) differences caused by transit is useful for identifying the value of transit access by groups. We will also define the calc_slope function which we will use for calculating and controlling for the ZHVI trend and upholding the parallel trends assumption of DiD estimation.

```python
zhvi_transit['log_ZHVI'] = np.log(zhvi_transit['ZHVI'])

# Identify treatment events
treatment_events = zhvi_transit[(zhvi_transit['score_diff'].notna()) & (zhvi_transit['score_diff'] != 0)]

# Loop through treatment events
results = []

def calc_slope(series):
    if len(series) < 7:
        return pd.Series([np.nan] * len(series), index=series.index)
    else:
        return series.rolling(window=7).apply(lambda x: (x[-1] - x[0]) / 6, raw=True)
```

```python
for idx, row in treatment_events.iterrows():
    treated_nhood = row['neighborhood']
    treatment_date = int(row['yymm'])
    change_direction = np.sign(row['score_diff'])

    min_yymm = 1206
    max_yymm = 2503

    nhood_df = zhvi_transit[zhvi_transit['neighborhood'] == treated_nhood].copy()
    nhood_df = nhood_df.sort_values('yymm')

    nhood_score_series = nhood_df[nhood_df['yymm'].astype(int) >= treatment_date].sort_values('yymm')
    new_score = row['trans_score']
    post_scores = nhood_score_series.head(3)['trans_score'] 

    if post_scores.nunique() > 1 or (post_scores != new_score).any():
        continue        
    
    previous_change = nhood_df[(nhood_df['yymm'].astype(int) < treatment_date) & (nhood_df['score_diff'] != 0)]
    if not previous_change.empty:
        min_pre_date = int(previous_change.iloc[-1]['yymm'])
    else:
        min_pre_date = shift_yymm(treatment_date, -12)

    max_post_date = shift_yymm(treatment_date, 12)
    max_post_date = min(max_post_date, max_yymm)
    min_pre_date = max(min_pre_date, min_yymm)

    analysis_period = zhvi_transit[(zhvi_transit['yymm'].astype(int) >= min_pre_date) & 
                                   (zhvi_transit['yymm'].astype(int) <= max_post_date)].copy()

    pre_score = row['trans_score'] - row['score_diff']

    def score_stable_in_window(g):
        scores = g[(g['yymm'].astype(int) >= min_pre_date) & (g['yymm'].astype(int) <= max_post_date)]['trans_score']
        return scores.nunique() == 1

    stable_nhoods = zhvi_transit.groupby('neighborhood').filter(score_stable_in_window)['neighborhood'].unique()

    control_scores = zhvi_transit[(zhvi_transit['neighborhood'].isin(stable_nhoods)) & 
                                  (zhvi_transit['yymm'].astype(int) == min_pre_date)].copy()

    try:
        pre_ZHVI = zhvi_transit[(zhvi_transit['neighborhood'] == treated_nhood) & 
                                (zhvi_transit['yymm'].astype(int) == min_pre_date)]['ZHVI'].values[0]
        treated_geom = zhvi_transit[(zhvi_transit['neighborhood'] == treated_nhood)].geometry.iloc[0]
    except IndexError:
        continue

    treated_geom_centroid = treated_geom.centroid
    control_scores_geom_centroids = control_scores.geometry.centroid

    control_scores['ZHVI_diff'] = abs(control_scores['ZHVI'] - pre_ZHVI)
    control_scores['spatial_dist'] = control_scores_geom_centroids.distance(treated_geom_centroid)

    control_scores = control_scores[control_scores['ZHVI_diff'] < pre_ZHVI * 0.1]

    if change_direction > 0:
        candidate_controls = control_scores[control_scores['trans_score'] <= pre_score]
    else:
        candidate_controls = control_scores[control_scores['trans_score'] >= pre_score]


    selected_nhoods = [treated_nhood] + candidate_controls['neighborhood'].tolist()
    sub_df = analysis_period[analysis_period['neighborhood'].isin(selected_nhoods)].copy()

    sub_df['treated'] = (sub_df['neighborhood'] == treated_nhood).astype(int)
    sub_df['post'] = (sub_df['yymm'].astype(int) >= treatment_date).astype(int)
    sub_df['treated_post'] = sub_df['treated'] * sub_df['post']
    sub_df['log_trend'] = sub_df.groupby('neighborhood')['log_ZHVI'].transform(calc_slope)

    sub_df['weight'] = sub_df['SizeRank'].max() - sub_df['SizeRank'] + 1

    treated_log_trend = sub_df[(sub_df['neighborhood'] == treated_nhood)]['log_trend'].mean()
    control_scores['log_trend'] = zhvi_transit[
    (zhvi_transit['neighborhood'].isin(control_scores['neighborhood'])) &
    (zhvi_transit['yymm'].astype(int) == min_pre_date)
    ].groupby('neighborhood')['log_ZHVI'].apply(calc_slope).reset_index(drop=True)

    control_scores['trend_diff'] = abs(control_scores['log_trend'] - treated_log_trend)
    
    candidate_controls = control_scores.sort_values(['ZHVI_diff', 'trend_diff', 'spatial_dist']).head(5)
    treated_post_slope = sub_df[(sub_df['neighborhood'] == treated_nhood) & (sub_df['post'] == 1)]['log_ZHVI'].diff().mean()

    treated_pre_slope = sub_df[(sub_df['neighborhood'] == treated_nhood) & (sub_df['post'] == 0)]['log_ZHVI'].diff().mean()

    if change_direction < 0 and treated_post_slope > 0:
        continue  

    if sub_df['treated'].sum() < 3 or sub_df.shape[0] < 20 or candidate_controls.shape[0] == 0:
        continue

    try:
        model = smf.wls('log_ZHVI ~ treated + post + treated_post + log_trend + C(neighborhood) + C(yymm)', 
                        data=sub_df, weights=sub_df['weight']).fit()

        pval = model.pvalues.get('treated_post', np.nan)
        if pval < 0.05:
            results.append({
                'neighborhood': treated_nhood,
                'treatment_yymm': treatment_date,
                'coef': model.params.get('treated_post', np.nan),
                'pval': pval,
                'n_obs': sub_df.shape[0],
                'direction': 'increase' if change_direction > 0 else 'decrease'
            })
    except Exception as e:
        print(f"Error in regression for neighborhood {treated_nhood}: {e}")

results_df = pd.DataFrame(results)
```
